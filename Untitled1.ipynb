```python
#for estimate of the region we could probe using a limited array of PMTs

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import lambertw
from scipy.special import erf

placeholder = 0.0
Mx=0
p=10
Mmax = []
Smin = [] #array of all sigma_x values
v=[]
for y in range(0,3):

    for i in range(1,p):
        sigmax = i*10**-(6-y)
    
        for j in range(0,6):
            vx = 170 + j*50 #setting macro speed
            z=2.5*lambertw(3.16*10**16*(sigmax)**3*(vx/250)**6)
        #z=10*lambertw(1.6*10**17*(sigmax)**2*(vx/250)**4)
            k=np.real(z)
            if k >20:
                k=20
            if k<1:
                k=0
            Mx = 1.92 * (k)**2
            if vx <230:
                Mx=Mx*(1-(-1.95313*10**9 + 1.95312*10**9*np.exp(-((-170 + vx)**2/62500)) + (1.17702*10**9 - 6.92365*10**6*vx)*erf(0.68 - 0.004*vx))*(2.407*10**-9))
            if vx >230:
                Mx=Mx*(1-(6.1501*10**8*np.exp((0.00544 - 0.000016*vx)*vx) - 8.37805*10**8*np.exp(-(((-460 + vx)*vx)/62500))+ 
                    6.1501*10**8*np.exp(-(((-340 + vx)*vx)/62500)) + (1.17702*10**9 - 6.92365*10**6*vx)*erf(0.68 - 0.004*vx) 
                       + (1.59244*10**9 - 6.92365*10**6 *vx)*erf((-230 + vx)/250))*(2.407*10**-9))
            if Mx > placeholder:
                placeholder = Mx
#         print(placeholder,Mx)
           
        Mmax.append(placeholder)
        Smin.append(sigmax)
        
print(Mmax,Smin)
```
