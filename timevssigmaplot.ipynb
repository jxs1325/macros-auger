```python
# graph for t_I0 vs \sigma_x, with time bins for Auger and JEM-EUSO.

from scipy.interpolate import interp1d
import io
import csv
import sys
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import scipy
import os
x=np.logspace(-9,3)
#Define Plot
fig = plt.figure()
ax = fig.add_subplot(111)

#X-axis
ax.set_xlabel('$\sigma_\t{X}$ [cm$^{2}$]',fontsize=12)
ax.set_xlim([x.min(),x.max()])

#Y-axis
ymin=10**-11
ymax=10**0
ax.set_ylim([ymin,ymax])
ax.set_ylabel('$t_{I0}$ [s]',fontsize=12)

ax.loglog(x, 20*x*np.exp(-0.1), 'k', markersize=4, color='g',lw=1,label='D = 1 km',zorder=0)
ax.loglog(x, 20*x*np.exp(-0.1)*0.01, 'k', markersize=4, color='g',lw=1,label='D = 1 km, $v_x = 25 kms^{-1}$',linestyle='--',zorder=0)
ax.loglog(x, 20*x*np.exp(-0.1)*4, 'k', markersize=4, color='g',lw=1,label='D = 1 km $v_x = 500 kms^{-1}$',linestyle='-.',zorder=0)
ax.loglog(x, 20*x*np.exp(-0.5), 'k', markersize=4, color='k',lw=1,label='D = 5 km',zorder=0)
ax.loglog(x, 20*x*np.exp(-2), 'k', markersize=4, color='c',lw=1,label='D = 20 km',zorder=0)
ax.loglog(x, 10**-4*x**0, 'k', markersize=4, color='b',lw=1,label='Auger bin time',zorder=0)
ax.loglog(x, 2.5*10**-3*x**0, 'k', markersize=4, color='r',lw=1,label='JEM-EUSO bin time',zorder=0)
leg = ax.legend(loc='lower right', shadow=True)

ax.grid(True)
plt.savefig('time_vs_crosssection.pdf')
plt.show()
```
