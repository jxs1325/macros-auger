```python
# graph of number of photons produced vs \sigma_x

from scipy.interpolate import interp1d
import io
import csv
import sys
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import scipy
import os
q=np.logspace(-11,2)
q1=np.logspace(-11,-7)
qintaug=np.logspace(-7,-6.046)
q2=np.logspace(-6.046,-1.204)#euso not saturated one time bin
q3=np.logspace(-1.204,-1.0)#euso not saturated multi time bin
q4=np.logspace(-1.0,2)#euso saturated multi time bin
q5=np.logspace(-6.046,-2.397)#auger not saturated
q6=np.logspace(-2.397,-1.204)#auger saturated but single time bin
q7=np.logspace(-1.204,2)#auger saturated and multiple bin times
#Define Plot
fig = plt.figure()
ax = fig.add_subplot(111)
Augerdet = 6.81*10**-8
JEMdet = 1.5*10**-11
#X-axis
ax.set_xlabel('$\sigma_\t{X}$ [cm$^{2}$]',fontsize=12)
ax.set_xlim([q.min(),q.max()])

#Y-axis
ymin=10**0
ymax=10**17.0
ax.set_ylim([ymin,ymax])
ax.set_ylabel('$N_{\gamma}$',fontsize=12)

ax.loglog(q1, 7*10**27*q1**2*Augerdet, 'k', markersize=4, color='g',lw=1,label='Auger freestreaming',zorder=0)
ax.loglog(q1, 1.5*10**24*q1**2*JEMdet, 'k', markersize=4, color='b',lw=1,label='JEMEUSO freestreaming',zorder=0)
ax.loglog(qintaug, 3.657*10**38*qintaug**3.531*Augerdet, 'k', markersize=4, color='g',lw=1,label='Auger int',zorder=0,linestyle='dashed')
ax.loglog(qintaug, 8.046*10**34*qintaug**3.531*JEMdet, 'k', markersize=4, color='b',lw=1,label='JEMEUSO int',zorder=0,linestyle='dashed')
ax.loglog(q5, 1.92*10**26*q5**1.5*Augerdet, 'k', markersize=4, color='g',lw=1,label='Auger BB one bin below saturation',zorder=0)
ax.loglog(q6, 8*10**23*q6**0.5*Augerdet, 'k', markersize=4, color='g',lw=1,label='Auger BB below saturation multi bin',zorder=0)
ax.loglog(q7, 1.2*10**22*q7**-1*Augerdet, 'k', markersize=4, color='g',lw=1,label='Auger BB above saturation multi bin',zorder=0)
ax.loglog(q2, 4.22*10**22*q2**1.5*JEMdet, 'k', markersize=4, color='k',lw=1,label='JEMEUSO BB below saturation one bin',zorder=0)
ax.loglog(q3, 6.59*10**20*q3**0*JEMdet, 'k', markersize=4, color='k',lw=1,label='JEMEUSO BB above saturation one bin',zorder=0)
ax.loglog(q4, 6.59*10**19*q4**-1*JEMdet, 'k', markersize=4, color='k',lw=1,label='JEMEUSO BB above saturation multi bin',zorder=0)

#leg = ax.legend(loc='lower right', shadow=True)

ax.grid(True)
plt.savefig('number_size_macro.pdf')
plt.show()
```
